```python
from sklearn.model_selection import train_test_split
```

# dropping the first column as it has unique names and isnt going to be useful while building model

```python
df=df.drop(columns='name',axis=1)
```

```python
X = df.drop('status',axis=1)
y = df['status']
```

```python
rs=1  ##common random state declared
```

```python
train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=.30, random_state=rs)
```

# Data has been split 70:30, 70% for training and 30% testing

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
dt_model=DecisionTreeClassifier(criterion='entropy',random_state=rs)
dt_model.fit(train_set, train_labels)
y_predict = dt_model.predict(test_set)
print("Training data score",dt_model.score(train_set , train_labels))
print("Testing data score",dt_model.score(test_set , test_labels))
print("---------------------")
print("confustion matrix")
print(metrics.confusion_matrix(y_predict,test_labels))
print("---------------------")
cm=pd.crosstab(y_predict,test_labels,rownames=['Predicted'],colnames=['Actuals'])
print("Cross tab")
print(cm)
```

# The model has overfitted and the accuracy is not upto the expected level. 

```python
best_train_score=0
best_test_score=0
best_train_max_depth=0
best_train_min_samples=0
best_test_max_depth=0
best_test_min_samples=0
for x in range(2,20):
    for y in range(2,20):
        reg_dt_model = DecisionTreeClassifier(criterion = 'entropy',max_depth=x,min_samples_leaf=y,random_state=rs)
        reg_dt_model.fit(train_set, train_labels)
        y_predict = reg_dt_model.predict(test_set)
        a=reg_dt_model.score(train_set,train_labels)
        b=reg_dt_model.score(test_set,test_labels)
        if best_train_score<a:
            best_train_score=a
            best_train_max_depth=x
            best_train_min_samples=y
        if best_test_score<b:
            best_test_score=b
            best_test_max_depth=x
            best_test_min_samples=y
print("Best train score =",best_train_score,"when Max depth = ",best_train_max_depth,"and Min sample leaf =",best_train_min_samples)
print("Best test score =",best_test_score,"when Max depth = ",best_test_max_depth,"and Min sample leaf =",best_test_min_samples)
```

```python
reg_dt_model = DecisionTreeClassifier(criterion = 'entropy',max_depth=best_test_max_depth,min_samples_leaf=best_test_min_samples,random_state=rs)
reg_dt_model.fit(train_set, train_labels)
y_predict = reg_dt_model.predict(test_set)
print("Training data score =",reg_dt_model.score(train_set,train_labels))
print("Testing data score=",reg_dt_model.score(test_set,test_labels))
print("---------------------")
print("confustion matrix")
print(metrics.confusion_matrix(y_predict,test_labels))
print("---------------------")
cm=pd.crosstab(y_predict,test_labels,rownames=['Predicted'],colnames=['Actuals'])
print("Cross tab")
print(cm)
```

# we are able to get better accuracy in general after tweaking the regularization parameters, also we are able to stop the model from over fitting. accuracy for actual true vs incorrectly predicted as false is still not as expected, as this is the defining metrics for this usecase.

```python
from sklearn.ensemble import RandomForestClassifier
b_ntrain_e=0
b_ntest_e=0
best_testscore=0
best_trainscore=0
for x in range(2,100):
    rfcl = RandomForestClassifier(n_estimators = x,random_state=rs)
    rfcl = rfcl.fit(train_set, train_labels)
    y_predict = rfcl.predict(test_set)
    a=rfcl.score(train_set , train_labels)
    b=rfcl.score(test_set , test_labels)
    if best_trainscore<a:
        b_ntrain_e=x
        best_trainscore=a
    if best_testscore<b:
        b_ntest_e=x
        best_testscore=b
print("Best train score =",best_trainscore,"when n_estimator = ",b_ntrain_e)
print("Best test score =",best_testscore,"when n_estimator = ",b_ntest_e)
```

```python
rfcl = RandomForestClassifier(n_estimators =b_ntest_e,random_state=rs)
rfcl = rfcl.fit(train_set, train_labels)
y_predict = rfcl.predict(test_set)
print("Training data score=",rfcl.score(train_set,train_labels))
print("Testing data score=",rfcl.score(test_set,test_labels))
print("---------------------")
print("confustion matrix")
print(metrics.confusion_matrix(y_predict,test_labels))
print("---------------------")
cm=pd.crosstab(y_predict,test_labels,rownames=['Predicted'],colnames=['Actuals'])
print("Cross tab")
print(cm)
```

# The model performs better in general and the important metric i.e false negative vs true postive is 0. Optimal number of tress  = 14
